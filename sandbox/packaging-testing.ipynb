# `mocksipipeline` Packaging Testing

```python
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt

from mocksipipeline.detector.response import Channel, SpectrogramChannel
from mocksipipeline.detector.filter import ThinFilmFilter
```

```python
al_thick = ThinFilmFilter(elements='Al',
                         thickness=200*u.nm,
                         xrt_table='Henke')
```

```python
chan_def = SpectrogramChannel(0)
chan_al = SpectrogramChannel(0, filter=al_thick)
```

```python
plt.plot(chan_def.wavelength, chan_def.wavelength_response)
plt.plot(chan_al.wavelength, chan_al.wavelength_response)
plt.yscale('log')
```
